```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
p_energy = 5.48956 * np.ones(10000) #MeV
```

```python
stop_power = 2.234e+02 #MeV cm2/g
rho_Au = 19.32 #g/cm3
thick_Au = 1.51e-04 # cm
sig_Au = rho_Au*thick_Au #g/cm2 (densità superificiale)
print(sig_Au)
```

```python
np_particles = 10000000
radius = 0.425
#0.335 #cm

p_energy = 5.48956 * np.ones(np_particles) #MeV

r_i = np.random.uniform(0,radius, np_particles)
r_f = np.random.uniform(0,radius, np_particles)
t_i = np.random.uniform(0,2*np.pi, np_particles)
t_f = np.random.uniform(0,2*np.pi, np_particles)

x_i = r_i*np.cos(t_i)
x_f = r_f*np.cos(t_f)
y_i = r_i*np.sin(t_i)
y_f = r_f*np.sin(t_f)
 
dist = np.sqrt(np.square(x_f-x_i)+np.square(y_f-y_i)+thick_Au**2)
sig_particle = rho_Au*dist #g/cm2 (densità superificiale)
sig_a_cazzo_de_cane = rho_Au*thick_Au/np.cos(np.random.uniform(0,np.pi/4, np_particles))
```

```python
energy_loss = stop_power*sig_a_cazzo_de_cane#sig_particle
energy_final = p_energy-energy_loss
plt.hist(sig_a_cazzo_de_cane)
plt.show()
plt.hist(energy_loss, bins=100)
plt.show()
plt.hist(energy_final, bins=100)
#plt.yscale('log')
plt.show()
```
